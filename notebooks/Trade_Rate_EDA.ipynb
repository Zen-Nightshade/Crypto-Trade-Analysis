# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
filepath = "../data/processed/trades_2025-10-01_to_2025-10-07_1min.csv"
```

```python
df = pd.read_csv(filepath)
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
```

```python
counts = df["trade_count"]
sizes = df["amount"] * df["price"]
```

# Ploting Functions

```python
def plotArray(counts, title, xlabel, ylabel):
    plt.figure(figsize=(18,5))
    plt.plot(counts.index, counts.values, marker='o')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
def autoCorrelationFunction(counts, title):
    plt.figure(figsize=(18,5))
    pd.plotting.autocorrelation_plot(counts)
    plt.axvline(x=300, color='r', linestyle='--', linewidth=2)
    plt.title(title)
    plt.show()
```

# Autocorrelarions Plots

```python
autoCorrelationFunction(counts, title= "Trade Rate Autocorrelation Plot")
autoCorrelationFunction(sizes, title ="Trade Volume Autocorrelation Plot")
```

# Trade Plot

```python
plotArray(counts, "Trades per Minute", "Minutes", "Number of Trades")
plotArray(sizes, "Trade Costs per Minute", "Minutes", "Cost of Trades")
```
