```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "../data/processed/trades_2025-10-01_to_2025-10-07_1min.csv"
```

```python
df = pd.read_csv(filepath)
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
```

```python
def plotArray(counts, title, xlabel, ylabel):
    plt.figure(figsize=(10,5))
    plt.plot(counts.index, counts.values, marker='o')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
def autoCorrelationFunction(counts):
    pd.plotting.autocorrelation_plot(counts)
    plt.figure(figsize=(15,5))
    plt.show()
```

```python
counts = df["trade_count"]
sizes = df["amount"] * df["price"]
autoCorrelationFunction(counts)
autoCorrelationFunction(sizes)
plotArray(counts, "Trades per Minute", "Minutes", "Number of Trades")
plotArray(sizes, "Trade Costs per Minute", "Minutes", "Cost of Trades")
```
